```python
import geocoder # import geocoder
latitude = []
longitude = []

for i in range(len(pcode2)):
    # initialize your variable to None
    lat_lng_coords = None

    # loop until you get the coordinates
    while(lat_lng_coords is None):
      #g = geocoder.google('{}, Toronto, Ontario'.format(pcode[0]))
      g=geocoder.arcgis('{}, Toronto, Ontario'.format(pcode2[i]))
      lat_lng_coords = g.latlng


    latitude.append(lat_lng_coords[0])
    longitude.append(lat_lng_coords[1])
```

```python
df_coords = pd.DataFrame({'Latitude':latitude,'Longitude' : longitude})
```

```python
frames = [df, df_coords]
result = pd.concat(frames, axis = 1)
```
